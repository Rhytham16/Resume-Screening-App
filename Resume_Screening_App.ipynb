**Cleaning Data**

1 URLs,
2 hashtags,
3 mentions,
4 special letters,
5 punctuations:

```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\S',' ',txt)
    cleanText=re.sub('RT|cc',' ',cleanText)
    cleanText = re.sub('@\S+',' ',cleanText)
    cleanText = re.sub('#\S+',' ',cleanText)
    cleanText=re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""),' ',cleanText)
    cleanText=re.sub(r'[^\x00-\x7f]',' ',cleanText)
    cleanText = re.sub('\s+',' ',cleanText)



    return cleanText
```

```python
cleanResume("my website is this http://helloworld and world")
```

```python
cleanResume("my website is this http://helloworld and world IT @gmail.com")
```

```python
cleanResume("my website is this and world ##### ")
```

```python
df['Resume'] = df['Resume'].apply(lambda x:cleanResume(x))
```

```python
df['Resume'][0]
```

 **Words Into Categorical Values**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category']=le.transform(df['Category'])
```

```python
df.Category.unique()
```

**Vactorization**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredText =tfidf.transform(df['Resume'])
```

```python
df
```

```python
requiredText
```

**Splitting**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(requiredText,df['Category'],test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

**Now let's train the model and print the the Classification repoort:**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score


clf=OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
ypred=clf.predict(X_test)
print(ypred)
print(accuracy_score(y_test,ypred))
```
